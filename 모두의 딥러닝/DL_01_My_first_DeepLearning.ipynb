# 최소제곱법

 목표 : 정확한 선을 긋기

 즉, 정확한 기울기와 정확한 y절편을 알아내면 됨
 - 최소제곱법을 적용하여 일차함수의 기울기와 y절편을 바로 구할 수 있음

독립변수가 공부한 시간x일때, x의 편차를 제곱해서 합한 값을 분모로 놓고,
x와 y의 편차를 곱해서 합한 값을 분자로 놓으면 기울기가 나옴

```python
# 회귀분석의 기초_가장 훌륭한 예측선 그리기
import pandas as pd
import numpy as np

# 공부한 시간과 점수 x,y 변수에 넣기
x = np.array([2,4,6,8]) # 공부한 시간
y = np.array([81,93,91,97]) # 시험 본 점수
```

```python
# x,y,의 평균 값 구하기
mx = np.mean(x)
my = np.mean(y)

print('x평균값: ', mx)
print('y평균값: ', my)
```

```python
[(i-mx)**2 for i in x]
```

```python
sum([(i-mx)**2 for i in x])
```

```python
# 기울기 공식  y = wx + b
# w(weight) 가중치 b(bias) 편차

# x = np.array([2,4,6,8])

# 최소제곱법 분모 만들기
divisor = sum([(i-mx)**2 for i in x])
# x의 각 원소와 x의 평균 값들의 차를 제곱
# 편차의 제곱 합
# 독립변수x를 기준으로 편차 제곱

# 최소제곱법 분자 만들기
def top(x,mx,y,my):
  d = 0
  for i in range(len(x)): # len(x) = n
    d += (x[i]-mx) * (y[i]-my)
  return d

dividend = top(x,mx,y,my)

print('분모: ',divisor)
print('분자: ', dividend)
```

```python
# 평균 제곱 오차 구하기

# 기울기
w = dividend / divisor

# y절편 b
b = my - (mx*w)
# y - wx = b

print('기울기 w', w)
print('y 절편 b', b)
print('가장 훌륭한 예측선은 %.2fx + %d 이다' %(w,b))
```

```python
# 평균 제곱 오차

# 임의의 가상의 기울기 값w, 절편 b 입력
w = 3
b = 76

x = np.array([2,4,6,8])   # 공부한 시간
y = np.array([81,93,91,97])   # 시험 본 점수
```

```python
# 예측함수 (pred = wx +b)
def predict(x):
  return w * x + b

# 예측값이 들어갈 빈 리스트 생성
result =[]

for i in range(len(x)):
  result.append(predict(x[i]))
  print('공부한 시간: %.2f, 실제점수: %.2f, 예측점수: %.2f'%(x[i], y[i], predict(x[i])))
```

```python
# 평균 제곱 오차 함수 이용, 각 y값 대입 >> 최종값 구하기

n = len(x)

def mse(y,y_pred):
  return (1/n) * sum((y-y_pred)**2)

# 평균제곱오차 출력
print('평균제곱오차: ' + str(mse(y, result)))
```
